To get a summary of the data you can use NumPy functions:
- `np.mean(column)`

```python
avg = np.mean(np_baseball[:,0])
print("Average: " + str(avg))

# Print median height
med = np.median(np.baseball[:, 0])
print("Median: " + str(med))

# Print out the standard deviation on height
stddev = np.std(np.baseball[:, 0])
print("Standard Deviation: " + str(stddev))

# Print out correlation between first and second column
corr =  np.corrcoef(np_baseball[:, 0], np_baseball[:, 1])
print("Correlation: " + str(corr))
```

## 3. Comparison Operators with NumPy

### 3.1. Comparison Operators
Remember that NumPy works by vectorization so if you compare it it returns the comparison to each element in the array.

```python
# Create arrays
import numpy as np
my_house = np.array([18.0, 20.0, 10.75, 9.50])
your_house = np.array([14.0, 24.0, 14.25, 9.0])

# my_house greater than or equal to 18
print(my_house >= 18)

# my_house less than your_house
print(my_house < your_house)
```

### 3.2. Logical Operators in NumPy Arrays
- NP arrays require functions instead of just `and` `or`, and `not`
   - `np.logical_and()`
   - `np.logical_or()`
   - `np.logical_not()`

```python
bmi = np.array([22.1, 21.8, 23.5])
np.logical_and(bmi > 21, bmi < 22)  # [True, True, False]
```

You can use the result to filter the array:

```python
bmi[np.logical_and(bmi > 21, bmi < 22)]  # [22.1, 21.8]
```

### 6.3. Iterating over DataFrames:
- **If** we treat DF as 2D NumPy array then we may proceed with:

```python
import pandas as pd
brics = pd.DataFrame({
    'country': ['Brazil', 'Russia', 'India', 'China', 'South Africa'],
    'capital': ['Brasilia', 'Moscow', 'New Delhi', 'Beijing', 'Pretoria'],
    'area': [8.516, 17.10, 3.287, 9.597, 1.221]
})

for rows in brics:
    print(rows)
```

You can see that it only returns the name of the columns.

To iterate over rows, we must use the `.iterrows()` method.
- `.iterrows()` return the index and the Series of a row (tuple).

```python
for index, rows in brics.iterrows():
    print(index)
    print(rows)
```

We already know that we can just select a specific data from a series with:<br>
`seriesName['indexName']`. We can use the same logic when looping to the dataframe.

```python
for lab, row in brics.iterrows():
    print(lab, row['capital'])
```

### 6.4. Adding New Column with Loop
- Remember that we add a new column to a DataFrame with this:

```python
df['new_col'] = array
```

We can use the same logic when looping:

```python
for lab, row in brics.iterrows():
    brics.loc[lab, 'name_length'] = len(row['country'])
print(brics)
```

**Explanation:**<br> You are iterating at every row (lab) and connecting it to the column (name_length).<br>For every iteration you are assigning a cell to the value returned by the len taken from a data from the series. **ANKI THIS**

### 6.5. More Efficient Method:
- The `apply()` function is more efficient for applying a function to an entire column element-wise.

```python
brics['name_length'] = brics['country'].apply(len)
print(brics)
```
