## Importing Flat Files with NumPy

We load flat files using numpy if most of data are numbers

## Functions

| Functions | Description | Syntax |
|---|---|---|
| `loadtxt` | Loads data from a text file into a NumPy array. | `numpy.loadtxt(filename, delimiter=None, skiprows=0, usecols=None, dtype=None)` |

## Arguments

| Arguments | Description | Syntax |
|---|---|---|
| `filename` | The name of the file to load. | `numpy.loadtxt(filename, ...)` |
| `delimiter` | The string used to separate values. | `numpy.loadtxt(..., delimiter=delimiter_string, ...)` |
| `skiprows` | The number of rows to skip at the beginning of the file. | `numpy.loadtxt(..., skiprows=n, ...)` |
| `usecols` | Which columns to read, with 0 being the first. | `numpy.loadtxt(..., usecols=[n, n], ...)` |
| `dtype` | The data type of the resulting array. | `numpy.loadtxt(..., dtype=data_type, ...)` |
