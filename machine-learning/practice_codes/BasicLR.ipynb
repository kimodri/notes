```python
# Import the necessary modules
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv(r"C:\Users\magan\Desktop\HousePrice.csv")
print(df)
```

```python
# get the data
x = np.array(df["SqFt"].tolist())
y = np.array(df["Price"].tolist())
```

```python
#plot the dataset
plt.title("House Prices")
plt.xlabel("Sq Feet")
plt.ylabel("Prices")
plt.grid()
plt.scatter(x, y)
plt.show()
```

```python
# create an object 
model = LinearRegression()

# use the obejct to fit or train the model
model.fit(x.reshape(-1, 1), y)
```

```python
# predicting price with another value
newHouseSize = np.array([[1600]]) #why list sa loob ng list?!
newPrice = model.predict(newHouseSize)
print(newPrice)
```

```python
plt.scatter(x, y, color = 'red', label = 'Training data')
plt.scatter(newHouseSize, newPrice, color = 'blue', label = 'Testing data')
xRange = np.linspace(min(x), max(x), 100) # i think this is optional
yRange = model.predict(xRange.reshape(-1, 1)) # what this needs remember is a 2d array
plt.plot(xRange, yRange, color = 'yellow', label = 'Regression Line')
plt.xlabel("Sq Feet")
plt.ylabel("Prices")
plt.legend()
plt.show()
```

```python
plt.scatter(x, y, color = 'red', label = 'Training data')
plt.scatter(newHouseSize.flatten(), newPrice, color = 'blue', label = 'Testing data')
xRange = np.linspace(min(x), max(x), 100) # i think this is optional
yRange = model.predict(xRange.reshape(-1, 1)) # what this needs remember is a 2d array
plt.plot(xRange, yRange, color = 'yellow', label = 'Regression Line')
plt.xlabel("Sq Feet")
plt.ylabel("Prices")
plt.legend()
plt.show()
```
