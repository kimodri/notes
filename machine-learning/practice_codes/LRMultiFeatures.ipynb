## Import the libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
```

## Import the data

```python
from sklearn import datasets
diabetes = datasets.load_diabetes()
df = pd.DataFrame(data = diabetes.data, columns = diabetes.feature_names)
df['target'] =  diabetes.target
print(df.head())
```

## Define x and y

```python
x = np.array(df.drop(columns = ['target']))
y = np.array(df['target'])
```

## Split the data

```python
xTrain, xTest, yTrain, yTest = train_test_split(x, y, random_state = 42, test_size = 0.2)
print("Training dataset's shape, testing dataset's shape: ", xTrain.shape, xTest.shape)
print("Training dataset's shape, testing dataset's shape: ", yTrain.shape, yTest.shape)
```

## Train data

```python
model = LinearRegression()
model.fit(xTrain, yTrain)
```

## Predict data
From the testing data

```python
predictedVal = model.predict(xTest)
```

## Plot the graph to see actual and predicted values

```python
plt.scatter(yTest, predictedVal, color = 'black')
plt.grid()
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.title("Correlation among Multiple Factors and Disease Progression (Diabetes)")
plt.show()
```

## Error analysis with r2
1.0 is the highest

```python
entirePredict = model.predict(x)
r2E = r2_score(y, entirePredict)
print(round(r2E, 2))
```

## Graph of the entire actual vs predicted values

```python
plt.scatter(y, entirePredict, color = 'black')
plt.grid()
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.title("Correlation among Multiple Factors and Disease Progression (Diabetes)")
plt.show()
```

## Realization
Indeed, that your model will predict better if you have larger datasets for when I did the models with just one feature such as with BP and Disease Progression, I only got an R2 score of 0.17. And with Sugar Level and Disase progression I got an R2 score of 0.11.<br>
But this time with multiple features, I got a score of 0.51.<br><br>
Nevertheless, there is still an insight, perhaps blood pressure is more correlated to disease progression than with sugar level.
